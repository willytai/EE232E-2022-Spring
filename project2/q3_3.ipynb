```r
df = read.table("cora/cora.content", header = FALSE, sep = "", dec = ".")
df = df[match(V(gcc)$name, df$V1),] #reorder the vertices to match the nodeID created by graph
row.names(df) = NULL
head(df)
# create feature matrix
feature_mat = data.matrix(df[, 2:1434])
colnames(feature_mat) = NULL
feature_mat
# randomly select 20 documents from each class as 'seeds'
classes = c('Case_Based', 
            'Genetic_Algorithms', 
            'Neural_Networks',
            'Probabilistic_Methods',
            'Reinforcement_Learning',
            'Rule_Learning',
            'Theory')
seeds = hash()
for(c in classes){
    v_ind= which(df$V1435 == c)
    #print(df[v_ind,]$V1435[1:2])
    #v_names = v_names[match(v_names, V(gcc)$name)]   
    selected = sample(v_ind, size=20) # randomly select 20 seeds 
    seeds[[c]] = c(selected)
}
print(seeds)
```

```r
create_cos_simiarity_matrix = function (feature_mat){
    return(cosine(t(feature_mat)))
}
create_transition_matrix = function (g, cos_mat){
    vs = V(g)
    n = vcount(g)
    adj = as_adjacency_matrix(g)
    #print('adjacency mat')
    #print(adj)
    cos_mat = adj*cos_mat
    cos_mat[diag(rowSums(cos_mat) == 0)] = 1  
    #print('cosine mat intermediate result')
    #print(cos_mat)
    #z = matrix(rowSums(cos_mat, , 1))
    exp_mat = exp(cos_mat)
    exp_mat[which(cos_mat == 0)] = 0
    z = matrix(rowSums(exp_mat, , 1))
    #transition_matrix = cos_mat / repmat(z, 1, n)
    transition_matrix = exp_mat / repmat(z, 1, n)
    return(transition_matrix)
}
```

```r
random_walk_custom = function (g, 
                                num_steps, start_node, 
                                transition_matrix = NULL, alpha=0.0, feature_mat,
                                visit_mode="seed_document", seeds_per_class){
    if(is.null(transition_matrix)){
            cos_mat = create_cos_simiarity_matrix(feature_mat)
            transition_matrix = create_transition_matrix(g, cos_mat) 
    }
    v = start_node
    visited_nodes = array(data=0.0, length(num_steps))
    if (visit_mode == "seed_document"){
        visit_probs=array(0, vcount(g))
        visit_probs[seeds_per_class]=1/20   
    }
    else{
        visit_probs=array(1/vcount(g), vcount(g))
    }
    for(i in 1:num_steps){
        if(runif(1)<alpha){
            v = sample(1:vcount(g), 1, prob = visit_probs)   
        }
        else{
            PMF = transition_matrix[v, ]
            if(sum(is.nan(PMF))!=0){
                print(paste(which(is.nan(PMF)),v))} 
            v = sample(1:vcount(g), 1, prob = PMF) 
        }
        visited_nodes[i] = v
    }  
    return(visited_nodes)
}
```

```r
cos_mat = create_cos_simiarity_matrix(feature_mat)
transition_matrix = create_transition_matrix(gcc, cos_mat)
```

```r
page_rank = function(gcc, transition_matrix, feature_mat, classes, seeds, alpha){
    visits_counted_by_class = list()
    #cos_mat = create_cos_simiarity_matrix(feature_mat)
    #transition_matrix = create_transition_matrix(gcc, cos_mat) 
    for(class in classes){
        num_steps = 1000
        visited_counts = array(data=0.0, vcount(gcc))
        stable_step = ceiling(log(vcount(gcc)))
        for(i in 1:100){
            start = sample(1:vcount(gcc), 1)
            visited_nodes = random_walk_custom(g = gcc,num_steps = num_steps,
                                                transition_matrix = transition_matrix
                                                ,start_node = start,
                                                alpha = alpha, feature_mat = feature_mat,
                                                visit_mode ='seed_document',
                                                seeds_per_class = seeds[[class]] )
            for (j in (stable_step+1):length(visited_counts)) {
                visited_counts[visited_nodes[j]] = visited_counts[visited_nodes[j]] + 1
            }
        }
        visits_counted_by_class = append(visits_counted_by_class, list(class=visited_counts))
    }
    return(visits_counted_by_class)
}
```

```r
library('plyr')
results = list()
for(alpha in c(0, 0.1, 0.2)){
    visits_counted_by_class = page_rank(gcc, transition_matrix, feature_mat, classes, seeds, alpha)
    visit_mat = matrix(unlist(visits_counted_by_class), ncol = 7, nrow = 2485)
    prediction = factor(argmax(visit_mat))
    #prediction
    gt_classname = df$V1435
    #classes
    #gt_classname
    gt = factor(strtoi(mapvalues(gt_classname, classes, c(1:7))))
    #gt
    print(paste('results of alpha=', alpha))
    result = confusionMatrix(data = prediction, reference = gt)
    f1 = mean(result$byClass[,'F1'])
    acc = result$overall['Accuracy']
    print(acc)
    print(f1)
    results = append(results, list(data = prediction, reference = gt))
}
```

```r
# evaluate only on unlabeled data
for(j in c(1,2,3)){
    for(i in seq_along(classes)){
        labeled_ind = seeds[[classes[i]]]
        pred = results[[(2*j-1)]]
        gt = results[[(2*j)]]
        pred = pred[-c(labeled_ind)]
        gt = gt[-c(labeled_ind)]
    }
    result = confusionMatrix(data = factor(pred), reference = factor(gt))
    f1 = mean(result$byClass[,'F1'])
    acc = result$overall['Accuracy']
    print(acc)
    print(f1)
}
```
